```python
# Create some data
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(
        data[col], 
        density=True, 
        stacked=True, 
        alpha=0.5
    )
    plt.rcParams["figure.figsize"] = (15, 10)
    plt.legend('XY', ncol=3, loc='upper left', prop={'size': 10});
```

```python
for col in 'xy':
    sns.kdeplot(data[col], shade=True, cbar=True)
    sns.kdeplot(data.x, color='red', fill=True, label='X', alpha=0.1)
    sns.kdeplot(data.y, color='blue', fill=True, label='Y', alpha=0.1)
```

```python
sns.distplot(data['x'])
sns.distplot(data['y']);
```

```python
sns.displot([data['x'], data['y']], height=15)
```

```python
sns.kdeplot(data.x, data.y);
```

```python
with sns.axes_style('darkgrid'):
    sns.jointplot("x", "y", data, kind='kde');
```

```python
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='hex')
```

# Tips

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.describe()
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
grid = sns.FacetGrid(tips, row="sex", col="time", palette='Set2', hue='sex', margin_titles=True, height=6)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
```

```python
sns.set(font_scale=2)
with sns.axes_style(style='ticks'):
    g = sns.catplot("day", "total_bill", "sex", data=tips, kind="box", height=15)
    g.set_axis_labels("Day", "Total Bill");
```

```python
with sns.axes_style('white'):
    sns.jointplot("total_bill", "tip", data=tips, kind='hex', height=15)
```

```python
sns.jointplot("total_bill", "tip", data=tips, kind='reg', height=15);
```

```python
fig, ax = plt.subplots()
sns.heatmap(tips.corr(), annot=True, linewidths=0.1, fmt='.3g')
plt.show()
```

```python
data = pd.read_csv('marathon-data.csv', converters={'split':pd.to_timedelta, 'final':pd.to_timedelta})
data.head()
```

```python
data.dtypes
```

```python
data['split_sec'] = data['split'].astype(np.int64) / 1E9
data['final_sec'] = data['final'].astype(np.int64) / 1E9
data.head()
```

```python
with sns.axes_style('whitegrid'):
    g = sns.jointplot("split_sec", "final_sec", data, kind='hex', height=15)
    g.ax_joint.plot(np.linspace(4000, 16000), np.linspace(8000, 32000), ':k')
    g.refline(x=7000, y=16000)
```

```python
data['split_frac'] = 1 - 2 * data['split_sec'] / data['final_sec']
data.head()
```

```python
sns.set(font_scale=1)
sns.displot(data['split_frac'], kde=False, height=15);
plt.axvline(0, color="k", linestyle="--");
```

```python
sum(data.split_frac < 0)
```

```python
sns.set(font_scale=1)
g = sns.PairGrid(data, vars=['age', 'split_sec', 'final_sec', 'split_frac'],
                 hue='gender', palette='RdBu_r', height=3)
g.map(plt.scatter, alpha=0.8)
g.add_legend();
```

```python
sns.set(font_scale=1.5)
sns.kdeplot(data.split_frac[data.gender=='M'], label='men', shade=True, color='lightblue', alpha=0.5)
sns.kdeplot(data.split_frac[data.gender=='W'], label='women', shade=True, color='lightpink', alpha=0.5)
plt.ylim(0, 10)
plt.xlim(-0.3, 0.65)
plt.xlabel('split_frac');
```

```python
sns.violinplot("gender", "split_frac", data=data,
               palette=["lightblue", "lightpink"]);
```

```python
data['age_dec'] = data.age.map(lambda age: 10 * (age // 10))
data.head()
```

```python
men = (data.gender == 'Men')
women = (data.gender == 'Women')

with sns.axes_style(style='darkgrid'):
    sns.violinplot("age_dec", "split_frac", hue="gender", data=data,
                   split=True, inner="quartile",
                   palette=["lightblue", "lightpink"]);
```
